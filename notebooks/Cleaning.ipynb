```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
input_path = PROJECT_ROOT / "Chagas_DB_3.xlsx"
output_path = PROJECT_ROOT / "Chagas_DB_3_sin_columnas.xlsx"

columns_to_remove = [
    "Tromboembolimo pulmonar", "AHA", "NYHA", "PCR", "LCR", "Tromboembolimo pulmonar",
    "SIDA", "Metástasis sólidas", "Hemiplejía", "Leucemia", "Enfermedad Ulcerativa péptica",
    "Hitoria FAMILIAR de coronariopatía", "Id base", "Fecha de ingreso", "Estado egreso",
    "último contacto"
]

def normalize_name(name: str) -> str:
    return " ".join(str(name).strip().split()).lower()

df = pd.read_excel(input_path)
existing_map = {normalize_name(col): col for col in df.columns}

resolved = []
missing = []
for col in columns_to_remove:
    key = normalize_name(col)
    if key in existing_map:
        resolved.append(existing_map[key])
    else:
        missing.append(col)

resolved_unique = sorted(set(resolved), key=lambda c: list(df.columns).index(c))
df_new = df.drop(columns=resolved_unique)
df_new.to_excel(output_path, index=False)

print(f"Archivo creado: {output_path}")
print(f"Columnas originales: {len(df.columns)}")
print(f"Columnas eliminadas: {len(resolved_unique)}")
print(f"Columnas finales: {len(df_new.columns)}")
print("\nColumnas eliminadas:")
for c in resolved_unique:
    print(f"- {c}")

if missing:
    print("\nNo encontradas:")
    for c in missing:
        print(f"- {c}")
```
